```python
#crop_im = full_im[290:660,600:1640]
# 1080*1920 original
# 370*1040 crop
# DEL x = 290 
#del y = 600

import pandas as pd
import numpy as np
import os
import pickle
```

```python
folder = "C:/Users/hp/Desktop/Research_intern_iitd/trajectories"
depth_path = "C:/Users/hp/Desktop/Research_intern_iitd/c2d"
cnt = 0
del_x = 290
del_y = 600
x_ratio_1 = 370/416
y_ratio_1 = 1040/416
x_ratio_2 = 451/1080
y_ratio_2 = 534/1920
for file in os.listdir(folder):
    path = folder + '/' + file
    df = pd.read_csv(path)
    x = []
    y = []
    r = []
    b = []
    g = []
    d = []
    for xc in df['raw_x']:
        if(xc==-1):
            x.append(-1)
            continue
        cur_x = ((x_ratio_1*xc)+del_x)*x_ratio_2
        x.append(cur_x)
        #print(int(cur_x))
    for yc in df['raw_y']:
        if(yc==-1):
            y.append(-1)
            continue
        cur_y = ((y_ratio_1*yc)+del_y)*y_ratio_2
        y.append(cur_y)
    for i in range(943):
        var=i
        suffix=str(var).zfill(4)
        print(file," : ",i)
        cur_file = pickle.load(open(r"C://Users//hp//Desktop//Research_intern_iitd//c2d//{}.txt".format(suffix),"rb"))
        np_file = np.array(cur_file)
        #print(np_file)
        row = int(x[i])
        col = int(y[i])
        if(row==-1):
            rgbd=[-1,-1,-1,-1]
        else:
            rgbd = np_file[row][col]
        r.append(rgbd[0])
        b.append(rgbd[1])
        g.append(rgbd[2])
        d.append(rgbd[3])
    df['x'] = x
    df['y'] = y
    df['d'] = d
    df['r'] = r
    df['g'] = g
    df['b'] = b
    df.to_csv("C:/Users/hp/Desktop/Research_intern_iitd/final_traj/{}".format(file))
```
